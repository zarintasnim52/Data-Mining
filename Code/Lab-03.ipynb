```python
dataset=[['milk','eggs','bread'],
        ['milk','bread'],
        ['eggs','sugar'],
        ['coffe','sugger'],
        ['onion','coffee','carrot','icecream']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder=TransactionEncoder().fit(dataset)
```

```python
onehot=encoder.transform(dataset)
```

```python
onehot=pd.DataFrame(onehot,columns=encoder.columns_)
```

```python
print(onehot)
```

```python
import scipy.spatial
```

```python
jaccard=scipy.spatial.distance.cdist(onehot,onehot,metric='jaccard')
```

```python
print(jaccard)
```

```python
jaccarddis=pd.DataFrame(jaccard,columns=onehot.index.values,index=onehot.index.values)
```

```python
print(jaccarddis)
```

```python
from scipy.spatial import distance
```

```python
mink_dist0 = distance.minkowski(jaccarddis[0],3)
mink_dist1 = distance.minkowski(jaccarddis[1],3)
mink_dist2 = distance.minkowski(jaccarddis[2],3)
mink_dist3 = distance.minkowski(jaccarddis[3],3)
mink_dist4 = distance.minkowski(jaccarddis[4],3)
print(mink_dist0)
print(mink_dist1)
print(mink_dist2)
print(mink_dist3)
print(mink_dist4)
```

0 is closer to everyone compared to others
